```python
# Import necessary libraries
import numpy as np
import pandas as pd
import os
import cv2
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Function to load images and labels
def load_data(data_dir):
    images = []
    labels = []
    for label in ["with_mask", "without_mask", "mask_weared_incorrect"]:
        path = os.path.join(data_dir, label)
        for img in os.listdir(path):
            img_path = os.path.join(path, img)
            image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            image = cv2.resize(image, (64, 64))  # Resize to 64x64
            images.append(image.flatten())  # Flatten the image
            labels.append(label)
    return np.array(images), np.array(labels)
```

```python
# Load data
data_dir = '/kaggle/input/face-mask-detection/images'  # Update the path
images, labels = load_data(data_dir)
```

```python
# Convert labels to numerical values
label_map = {'with_mask': 0, 'without_mask': 1, 'mask_weared_incorrect': 2}
labels = np.array([label_map[label] for label in labels])

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)
```

```python
# Feature extraction using CountVectorizer
vectorizer = CountVectorizer(analyzer='char', binary=True)
X_train_counts = vectorizer.fit_transform(X_train.astype('str'))
X_test_counts = vectorizer.transform(X_test.astype('str'))

# Train Multinomial Naive Bayes model
model = MultinomialNB()
model.fit(X_train_counts, y_train)
```

```python
# Detailed training process
# 1. Initialize Multinomial Naive Bayes model
model = MultinomialNB()

# 2. Fit the model with training data
model.fit(X_train_counts, y_train)

# 3. Make predictions on training data
train_predictions = model.predict(X_train_counts)

# 4. Evaluate training performance
train_accuracy = accuracy_score(y_train, train_predictions)
print(f"Training Accuracy: {train_accuracy * 100:.2f}%")

# 5. Make predictions on test data
y_pred = model.predict(X_test_counts)

```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

# Additional evaluation metrics
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=['with_mask', 'without_mask', 'mask_weared_incorrect']))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
